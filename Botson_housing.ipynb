```python
import pandas as pd
import numpy as np

# Load the dataset (from cell 2906b2b1)
df = pd.read_csv("/content/HousingData.csv")

# Display the first few rows and data types (from cell 60e39175)
display(df.head())
display(df.info())

# Check for and remove duplicate rows (from cell 60e39175)
initial_duplicates = df.duplicated().sum()
print(f"Number of duplicate rows before removal: {initial_duplicates}")
df.drop_duplicates(inplace=True)
duplicates_after_removal = df.duplicated().sum()
print(f"Number of duplicate rows after removal: {duplicates_after_removal}")


missing_values = df.isnull().sum()
print("Columns with missing values and their counts:")
print(missing_values[missing_values > 0])

# Explore columns with missing values: CRIM, ZN, INDUS, CHAS, AGE, LSTAT
# For numerical columns, checking value counts might not be very insightful unless there are specific sentinel values.
# For CRIM, ZN, INDUS, AGE, LSTAT, we will impute with the median as they are numerical and the median is robust to outliers.
# For CHAS, it's a binary variable (0 or 1). We'll impute with the mode.

# Impute missing values
for col in ['CRIM', 'ZN', 'INDUS', 'AGE', 'LSTAT']:
    df[col].fillna(df[col].median(), inplace=True)

df['CHAS'].fillna(df['CHAS'].mode()[0], inplace=True)

print("\nMissing values after imputation:")
print(df.isnull().sum())

# Calculate IQR and bounds for numerical columns (from cell c2268a1f)
numerical_cols = df.select_dtypes(include=np.number).columns
outlier_bounds = {}
for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outlier_bounds[col] = (lower_bound, upper_bound)

# Cap outliers at the calculated bounds (from cell)
df_capped = df.copy()
for col, bounds in outlier_bounds.items():
    lower_bound, upper_bound = bounds
    df_capped[col] = np.where(df_capped[col] < lower_bound, lower_bound, df_capped[col])
    df_capped[col] = np.where(df_capped[col] > upper_bound, upper_bound, df_capped[col])

# Scale the numerical data (from cell 75fbf5f0)
from sklearn.preprocessing import StandardScaler
numerical_cols_capped = df_capped.select_dtypes(include=np.number).columns
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df_capped[numerical_cols_capped])
df_scaled = pd.DataFrame(scaled_data, columns=numerical_cols_capped)
```
